```python
# feature engineering
# extracts the day from first part of 'Date_of_Journey' column an assigns it to 'Date' column
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
```

```python
# extracts the month from second part of 'Date_of_Journey' column 
# assigns it to 'month' column
df['month']=df['Date_of_Journey'].str.split('/').str[1]
```

```python
# extracts the year from third part of 'Date_of_Journey' column
# assigns it to 'year' column
df['year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
# display a quick overview of the first 5 rows
df.head()
```

```python
df.info()
```

```python
# converting date, month, year from object type to int type
df['Date'] = df['Date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)
```

```python
df.info()
```

```python
# dropping the 'Date_of_Journey' column
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
# applies a function that splits each value on a space (' ') in Arrival_Time column
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
# splits 'Arrival_hour' and 'Arrival_min' columns with a colon from index 0 and index 1
# assigns both columns from 'Arrival_Time' 
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
# converting 'Arrival_hour' and 'Arrival_min' to int
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
# splits 'Deep_hour' and 'Dep_min' columns with a colon from index 0 and index 1
# assigns both columns from 'Dep_Time'
df['Deep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df.info()
```

```python
# returns unique values in column 'Total_Stops'
df['Total_Stops'].unique()
```

```python
# returns counts of unique values in column 'Total_Stops'
df['Total_Stops'].value_counts()
```

```python
# filters rows with missing values in column 'Total_Stops'
df[df['Total_Stops'].isnull()]
```

```python
# returns frequent values
df['Total_Stops'].mode()
```

```python
# maps categorical values in 'Total_Stops' column to numerical values 
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2,
                                          '3 stops':3,'4 stops':4, np.nan:1})
```

```python
# filters to see which rows have missing NaNs
df[df['Total_Stops'].isnull()]
```

```python
df.head(2)
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
# extracts hours and min in 'Duration' column 
# splits 'Duration_hours' and 'Duration_min' column with ' '
# further splits 'Duration_hours' and 'Duration_min' with 'h' and 'm'
df['Duration_hours']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head(2)
```

```python
df.drop('Duration', axis=1, inplace=True)
```
